Verify evidence fusion and probability calculation.

```python
from sherlock_candidate_identifier.confidence import ConfidenceEngine
from sherlock_candidate_identifier.models import EvidenceType, MeetingMetadata, MeetingState, Participant, SignalResult

state = MeetingState(meeting_id='nb', metadata=MeetingMetadata(), participants={'a': Participant(participant_id='a', display_name='Candidate'), 'b': Participant(participant_id='b', display_name='Interviewer')})
results = [SignalResult(participant_id='a', signal_name='demo', evidence_type=EvidenceType.METADATA, score=1, confidence=1, explanation='candidate email match'), SignalResult(participant_id='b', signal_name='demo', evidence_type=EvidenceType.METADATA, score=-1, confidence=1, explanation='known interviewer')]
ConfidenceEngine().predict(state, results, 'demo explanation').model_dump()
```
