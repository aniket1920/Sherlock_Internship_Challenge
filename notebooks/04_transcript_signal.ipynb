```python
from sherlock_candidate_identifier.models import MeetingMetadata, MeetingState, Participant
from sherlock_candidate_identifier.signals import TranscriptSignal

state = MeetingState(meeting_id='nb', metadata=MeetingMetadata())
candidate = Participant(participant_id='candidate', display_name='Guest', transcript_text=['I built a fraud service and my project used streaming.'])
interviewer = Participant(participant_id='interviewer', display_name='Priya', transcript_text=['Tell me about your last project.'])
[TranscriptSignal().evaluate(p, state).model_dump() for p in [candidate, interviewer]]
```
