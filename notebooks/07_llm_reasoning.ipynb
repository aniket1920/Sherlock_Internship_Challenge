```python
from sherlock_candidate_identifier.models import (
    MeetingMetadata,
    MeetingState,
    Participant,
)
from sherlock_candidate_identifier.signals import LLMSignal

state = MeetingState(
    meeting_id="nb",
    metadata=MeetingMetadata(
        candidate_name="Nisha Rao",
        candidate_email="nisha@example.com",
        interviewer_names=["Priya Shah"],
        calendar_title="Technical Interview - Nisha Rao",
    ),
)

participant = Participant(
    participant_id="candidate",
    display_name="MacBook Pro",      # Device name instead of real name
    email="",                        # No email available
    camera_on=True,
    face_count=1,
    webcam_activity_score=0.7,
    speaking_duration_seconds=120,
    transcript_text=[
        "I built a YOLOv8 weed detection system.",
        "I trained the model on a Kaggle dataset.",
        "I implemented the Flask backend and integrated the detection pipeline.",
    ],
)

LLMSignal().evaluate(participant, state).model_dump()
```
